# Master DataFrame (Final Pipeline Concept)

Pipline/ Chaining (Understanding data Format and Build Pipeline of that)

Customers --> Orders (Order_items,reviews,payemnts [For Debugging Choose those one that have relationship with others]) --> Order_item --> products --> Products_Category --> sellers --> reviews --> Payments --> Geolocation                        

```python
# 1. Base Entity
master_df=customer_df
```

```python
# 2. orders
master_df=pd.merge(
    master_df,
    orders_df,
    on='customer_id',
    how='left'
)
```

```python
# 3. orders_items
master_df=pd.merge(
    master_df,
    order_items_df,
    on='order_id',
    how='left'
)
```

```python
# 4. Products
master_df=pd.merge(
    master_df,
    products_df,
    on='product_id',
    how='left'
)
```

```python
# 5. Products Category
master_df=pd.merge(
    master_df,
    product_category_df,
    on='product_category_name',
    how='left'
)
```

```python
# 6. Sellers
master_df=pd.merge(
    master_df,
    seller_df,
    on='seller_id',
    how='left'
)
```

```python
# Remove useless column BEFORE merging
master_df = master_df.drop(columns=["Unnamed: 0"], errors="ignore")
order_reviews_df = order_reviews_df.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
# 7. Reviews
master_df=pd.merge(
    master_df,
    order_reviews_df,
    on='order_id',
    how='left'
)
```

```python
# 8. Payments
master_df=pd.merge(
    master_df,
    order_payments_df,
    on='order_id',
    how='left'
)
```

```python
# Remove useless column BEFORE merging
geolocation_df = geolocation_df.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
# 9. Geolocation
master_df=pd.merge(
    master_df,
   geolocation_df,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)
```

```python
master_df.columns
```

# MASTER DATASET FIX

```python
# 1. REMOVE UNNAMED COLUMNS
master_df = master_df.loc[:, ~master_df.columns.str.contains("^Unnamed")]
```

```python
# 2. HANDLE DUPLICATE REVIEW COLUMNS
# (keep one clean review_score column)

master_df["review_score"] = master_df["review_score_x"].fillna(master_df["review_score_y"])

master_df.drop(columns=[
    "review_score_x",
    "review_score_y"
], inplace=True)
```
